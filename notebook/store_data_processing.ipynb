```python


# Handling CompetitionDistance (impute with 0)
sales['CompetitionDistance'].fillna(0, inplace=True)

# Handling CompetitionOpenSinceMonth and CompetitionOpenSinceYear (impute with median)
sales['CompetitionOpenSinceMonth'].fillna(sales['CompetitionOpenSinceMonth'].median(), inplace=True)
sales['CompetitionOpenSinceYear'].fillna(sales['CompetitionOpenSinceYear'].median(), inplace=True)

# Handling Promo2SinceWeek and Promo2SinceYear (impute with median)
sales['Promo2SinceWeek'].fillna(sales['Promo2SinceWeek'].median(), inplace=True)
sales['Promo2SinceYear'].fillna(sales['Promo2SinceYear'].median(), inplace=True)

# Handling PromoInterval (impute with a placeholder)
sales['PromoInterval'].fillna('NoPromo', inplace=True)
```
